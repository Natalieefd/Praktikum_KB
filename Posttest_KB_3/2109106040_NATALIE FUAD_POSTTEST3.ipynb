```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head(15)
```

### Gunakan minimal 3 tipe visualiasi data
##### Grafik Batang, Grafik Pie, Heatmap, dan Boxplot

##### ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
### Grafik Batang menggunakan Seaborn

### Menampilkan data dari atribut race/ethnicity dan atribut gender
##### Deskripsi : Menampilkan berapa banyak gender female dan male(object) pada setiap group(object) meggunakan grafik batang

```python
plt.figure(figsize=(10,5))

plt.title('Diagram race/ethnicity Dengan Kategori Gender')

sns.countplot(
    data=df,
    x="race/ethnicity",
    palette="BrBG",
    hue="gender"
)
plt.show()
```

### Menampilkan data dari atribut gender dan atribut lunch
##### Deskripsi : Menampilkan berapa banyak gender female dan male(object) yang makan siangnya gratis/dikurangi atau standar(object) menggunakan grafik batang

```python
plt.figure(figsize=(10,5))

plt.title('Diagram Gender Dengan Kategori Lunch')

sns.countplot(
    data=df,
    x="gender",
    palette="PRGn",
    hue="lunch"
)
plt.show()
```

### Menampilkan data dari atribut parental level of education dan atribut gender
##### Deskripsi : Menampilkan berapa banyak tingkat pendidikan orang tua/Parental Level Of Education(object) setiap siswa/i(gender) dengan berbagai kategori lulusan

```python
plt.figure(figsize=(10,5))

plt.title('Diagram Parental Level Of Education Dengan Kategori Gender')

sns.countplot(
    data=df,
    x="parental level of education",
    palette="RdYlGn",
    hue="gender"
)
plt.show()
```

### Grafik Batang menggunakan Matplotlib

### Menampilkan data dari atribut race/ethnicity dan atribut writing score
##### Deskripsi : Menampilkan berapa banyak siswa/i(gender) pada setiap grup yang memiliki skor menulis(writing score) sampai dengan 100 

```python
plt.figure(figsize=(5,8))

plt.title('Diagram Writing Score')
plt.xlabel('race/ethnicity')
plt.ylabel('writing score')

x = list(df["race/ethnicity"])
y = list(df["writing score"])

plt.bar(x, y, color="lightblue")
plt.show()
```

### Menampilkan data dari atribut race/ethnicity dan atribut reading score
##### Deskripsi : Menampilkan berapa banyak siswa/i(gender) pada setiap grup yang memiliki skor membaca(reading score) sampai dengan 100 

```python
plt.figure(figsize=(6,3.5))

plt.title('Diagram Reading Score')
plt.xlabel('race/ethnicity')
plt.ylabel('reading score')

x = list(df["race/ethnicity"])
y = list(df["reading score"])

plt.barh(x, y, color="lightblue")
plt.show()
```

##### ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
## Grafik Pie

```python
data_pie = df["math score"].value_counts()
data_pie.head()
```

```python
data_pie = df["math score"].value_counts().rename_axis("math score")
data_pie.head()
```

```python
data_pie = df["math score"].value_counts().rename_axis("math score").reset_index(name = "id_count")
data_pie.head()
```

### Menampilkan grafik pie dari atribut math score
#### Deskripsi : Menampilkan berapa persen setiap math score

```python
plt.figure(figsize=(15,15))

plt.pie(
    data_pie["math score"],
    labels = data_pie["math score"],
    autopct="%.2f%%"
)
plt.show()
```

##### ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
### Heatmap

### Menampilkan nilai korelasi dari data StudentsPerformance.csv
##### Deskripsi : Nilai dari atribut writing score merupakan nilai tertinggi, dan nilai dari atribut math score merupakan nilai paling kecil

```python
fig, ax = plt.subplots(figsize=(10,7))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap='PRGn'
)

ax.set_title('Korelasi antar atribut pada file StudentsPerformance.csv')
plt.show()
```

##### ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
## Boxplot

### Menampilkan nilai rata-rata dari atribut writing score
##### Deskripsi : (!perkiraan!)
#####                    - Pada atribut female memiliki nilai maximum 100, Q1-nya 81, median/Q2-nya 74, Q3-nya 62, nilai minimum 39, dan outlier-nya sebanyak 8.
#####                    - Pada atribut male memiliki nilai maximum 100, Q1-nya 74, median/Q2-nya 62, Q3-nya 50, nilai minimum 21, dan outlier-nya sebanyak 2.

```python
plt.figure(figsize = (5,9))

plt.title("Box Plot Rata-rata Writing Score Dengan Kategori Gender")

sns.boxplot(
    data = df,
    x = "gender",
    y = "writing score"
)

plt.show()
```

### Menampilkan nilai rata-rata dari atribut reading score
##### Deskripsi : (!perkiraan!)
#####                    - Pada atribut group B memiliki nilai maximum 99, Q1-nya 80, median/Q2-nya 64, Q3-nya 58, nilai minimum 23, dan outlier-nya tidak ada.
#####                    - Pada atribut group C memiliki nilai maximum 100, Q1-nya 79, median/Q2-nya 51, Q3-nya 60, nilai minimum 34, dan outlier-nya sebanyak 3.
#####                    - Pada atribut group A memiliki nilai maximum 100, Q1-nya 57, median/Q2-nya 63, Q3-nya 57, nilai minimum 23, dan outlier-nya tidak ada.
#####                    - Pada atribut group D memiliki nilai maximum 100, Q1-nya 79, median/Q2-nya 51, Q3-nya 60, nilai minimum 34, dan outlier-nya sebanyak 1.
#####                    - Pada atribut group E memiliki nilai maximum 100, Q1-nya 84, median/Q2-nya 77, Q3-nya 63, nilai minimum 34, dan outlier-nya sebanyak 1.

```python
plt.figure(figsize = (25,25))

plt.title("BoxPlot Rata-rata Reading Score Dengan Kategori race/ethnicity")

sns.boxplot(
    data = df,
    x = "race/ethnicity",
    y = "reading score"
)

plt.show()
```
